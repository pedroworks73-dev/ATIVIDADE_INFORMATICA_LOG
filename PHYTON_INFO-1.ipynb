```python
import pandas as pd

url = 'https://raw.githubusercontent.com/ybifoundation/Dataset/main/Salary%20Data.csv'
salary_data = pd.read_csv(url)

print(salary_data.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.regplot(x='Experience Years', y='Salary', data=salary_data, scatter_kws={'alpha':0.6}, line_kws={'color':'red'})
plt.title('Grfico de Regresso Linear: Anos de Experincia vs. Salrio')
plt.xlabel('Anos de Experincia')
plt.ylabel('Salrio')
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Prepare the data
X = salary_data[['Experience Years']].values # Features
y = salary_data['Salary'].values          # Target

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Get the coefficients
angular_coefficient = model.coef_[0]
linear_coefficient = model.intercept_

print(f"Coeficiente Angular (Inclinao): {angular_coefficient:.2f}")
print(f"Coeficiente Linear (Intercepto): {linear_coefficient:.2f}")
```

Os valores acima representam a equao da linha de regresso linear, onde:

*   **Coeficiente Angular (Inclinao)**: Indica o quanto o 'Salrio' (y) muda para cada unidade de aumento nos 'Anos de Experincia' (x). Neste caso, para cada ano adicional de experincia, o salrio aumenta em aproximadamente R$ {angular_coefficient:.2f}.
*   **Coeficiente Linear (Intercepto)**: Representa o valor esperado do 'Salrio' (y) quando os 'Anos de Experincia' (x) so zero. Em outras palavras,  o salrio base estimado.
